Load Dictionary yang sudah dibangun saat preprocessing

```python
# importing the module
import json
 
# Opening JSON file
def loadJSON(filename):
  data = {}
  with open(filename) as json_file:
      data = json.load(json_file)
  return data
```

```python
idx2tag = loadJSON('/content/drive/MyDrive/Colab Notebooks/FP-KK/output/idx2tag2.json')
tag2idx = loadJSON('/content/drive/MyDrive/Colab Notebooks/FP-KK/output/tag2idx2.json')
token2idx = loadJSON('/content/drive/MyDrive/Colab Notebooks/FP-KK/output/token2idx2.json')
idx2token = loadJSON('/content/drive/MyDrive/Colab Notebooks/FP-KK/output/idx2token2.json')
print(idx2tag)
```

Load Matrix hasil training dari Model dengan dataset baseline

```python
keras_model = tf.keras.models.load_model('/content/drive/MyDrive/Colab Notebooks/FP-KK/model/BiLSTM_baseline_fix.h5')
keras_weights = keras_model.get_layer("embedding").get_weights()[0]
```

```python
keras_weights[0]
```

```python
embedding_df = pd.DataFrame(keras_weights)
embedding_df
```

```python
dataset_new = dataset.drop(['Unnamed: 0', 'Word', 'Tag'], axis=1)
dataset_new
```

```python
keras_weights_dict = {}
for i in range(0, len(keras_weights)):
  keras_weights_dict[i] = keras_weights[i]
```

```python
dataset_new["embedding"] = dataset_new['Word_idx'].map(keras_weights_dict)
dataset_new
```

```python
embedding_series = dataset_new['embedding'].to_list()
embedding_df = pd.DataFrame(embedding_series)
```

```python
embedding_df
```

```python
embedding_df["Sentence #"] = dataset_new["Sentence #"]
embedding_df["Word_idx"] = dataset_new["Word_idx"]
embedding_df["Tag_idx"] = dataset_new["Tag_idx"]
embedding_df["Class"] = dataset_new["Class"]
embedding_df
```

```python
X = embedding_df.drop(['Class'], axis=1)
X
```

```python
y = embedding_df["Class"]
y
```

```python
from collections import Counter
from imblearn.over_sampling import SMOTENC
```

```python
print(f'Original dataset shape {X.shape}')
print(f'Original dataset samples per class {Counter(y)}')
```

```python
sm = SMOTENC(random_state=42, categorical_features=[100, 101, 102], sampling_strategy=0.5)
X_res, y_res = sm.fit_resample(X, y)
```

```python
print(f'Resampled dataset samples per class {Counter(y_res)}')
```
